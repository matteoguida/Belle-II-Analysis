## Some problems (?) with Signal events
- After filtering events for *'B0_isSignal'==1* we should only be left with correctly reconstructed B0 mesons. 
- If I am not mistaken 'Signal' events were created requiring that one side decays in our channel and the other side generically. 
- Given the low branching ratio of our decay channel we would expect close to zero events where both B0 are a Signal
- After filtering we are left with 242 rows with a non unique event number, suggesting that 121 events had two candidates qualifying as signal. 

```python
import pandas as pd
import numpy as np
from root_pandas import read_root
import matplotlib.pyplot as plt
```

```python
Signal = read_root('../DATA/B0_etapr-eta-gg2pi_KS-pi+pi-_output_sgnMC9_BGx1_0.root', 'B0')
```

```python
Signal = Signal[Signal['B0_isSignal']==1]
Duplicated = Signal.duplicated(subset='evt_no', keep=False)
Doubles = Signal[Duplicated]
Doubles = Doubles.reset_index(drop=True)
Doubles
```

```python
Masses = ['B0_M', 'B0_ErrM', 'B0_SigM', 'B0_K_S0_M', 'B0_K_S0_ErrM', 'B0_K_S0_SigM', 'B0_etap_M', 'B0_etap_ErrM', 
          'B0_etap_SigM', 'B0_etap_eta_M', 'B0_etap_eta_ErrM', 'B0_etap_eta_SigM']
Doubles[Masses]
```

```python
Doubles[Doubles.duplicated(subset='B0_M', keep=False)]
```

The reconstructed B0 candidates always have a different mass, this however is not true for the reconstructed K_S mass or the eta' mass:  

```python
Doubles[Doubles.duplicated(subset=['B0_K_S0_M', 'B0_K_S0_ErrM', 'B0_K_S0_SigM'], keep=False)]
```

```python
Doubles[Doubles.duplicated(subset=['B0_etap_M', 'B0_etap_ErrM', 'B0_etap_SigM'], keep=False)]
```

The last two lists are complementary. All events have either the same reconstructed K_S or eta' (Mass, ErrM and SigM are equal). No event exists where both K_S and eta' are the same.
